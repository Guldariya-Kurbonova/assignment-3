```python
import folium
```

```python
#create map object
#different tiles (map styles) can be used, like 'Stamen Toner', 'Stamen Terrain', ...
m = folium.Map(location=[51.177601, 71.432999], tiles='Stamen Terrain', zoom_start=4)
m
```

```python
#create city markers and add them to map object
folium.Marker(location=[42.333309, 69.621811],popup='SHYMKENT',tooltip='Click for more infromation', icon=folium.Icon(icon='cloud')).add_to(m)
folium.Marker(location=[43.222015, 76.851250],popup='ALMATY',tooltip='Click for more infromation', icon=folium.Icon(color='green')).add_to(m)
folium.Marker(location=[51.160522, 71.470360],popup='NUR-SULTAN',tooltip='Click for more infromation', icon=folium.Icon(color='red', icon='envelope')).add_to(m)
m
```

```python
#generate map and save as local file
m.save('index.html')
m
```
